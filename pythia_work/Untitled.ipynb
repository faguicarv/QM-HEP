```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
data = pd.read_csv("tomography_data.csv")

df_tau = data[data['id'] == 15].set_index('event')
df_antitau = data[data['id'] == -15].set_index('event')
df_higgs = data[data['id'] == 25].set_index('event')

hijos_sum = df_tau[['px', 'py', 'pz', 'E']] + df_antitau[['px', 'py', 'pz', 'E']]
diferencia = df_higgs[['px', 'py', 'pz', 'E']] - hijos_sum
```

```python
print(hijos_sum)
print(diferencia)
```

```python
E_total = hijos_sum['E']
P2_total = hijos_sum['px']**2 + hijos_sum['py']**2 + hijos_sum['pz']**2

m_reconstruida = np.sqrt(E_total**2 - P2_total)

print(f"Masa promedio reconstruida: {m_reconstruida.mean():.2f} GeV/c^2")
```

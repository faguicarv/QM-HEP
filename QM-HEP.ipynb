# Quantum tomography at High-energy physics service

In the following Notebook, we'll describe and implement a novel technique based on the work $\textit{Quantum tomography with } \tau $ $\textit{leptons at the FCC-ee: Entanglement, Bell inequality violation, sin } \theta_W$, $\textit{and anomalous couplings. M. Fabbrichesi and L. Marzola, Phys. Rev. D } \textbf{110}, 076004 (2024)$

A two spin-1/2 system is a way to study the two-qubit system (two-level system). In this direction, if we have a spin-1/2 particle in an arbitrary state then is usual to express the wave function as:

\begin{equation}
        \ket{\psi} = a_1 \ket{\uparrow} + a_2 \ket{\downarrow}
\end{equation}

In a two spin-1/2 system, then we have one particle in the Hilbert's space $\mathcal{H}_1$, and another particle in $\mathcal{H}_2$. So, it is important to note that the bipartite Hilbert's space will be denoted as $\mathcal{H} = \mathcal{H}_1 \otimes \mathcal{H}_2$. Then, our wave-function for the two spin-1/2 system is:

\begin{equation}
        \begin{split}
                \ket{\Psi} & = \ket{\psi_1} \otimes \ket{\psi_2} \\
                & = \left( a_1 \ket{\uparrow} + a_2 \ket{\downarrow} \right) \otimes \left( b_1 \ket{\uparrow} + b_2 \ket{\downarrow}\right) \\
                & = \left( a_1 b_1 \ket{\uparrow \uparrow} + a_1 b_2 \ket{\uparrow \downarrow} + a_2 b_1 \ket{\downarrow \uparrow} + a_2 b_2 \ket{\downarrow \downarrow}\right)
        \end{split}
\end{equation}

Also is possible to write the wave function as a matrix representation:
\begin{equation}
        \ket{\Psi} = \left[
                \begin{matrix}
                        a_1b_1 & a_1b_2 \\
                        a_2b_1 & a_2b_2 \\
                \end{matrix}
\right]
\end{equation}

But, in the last formalism we are asuming that the states are separable $\left(\ket{\Psi} = \ket{psi_1}\otimes\ket{\psi_2}\right)$, this means that the correlation between the two particles are not bound states. In order to allow the mixture between the two particles, we need to see if the non-diagonal elements are non-zero values. 

For this matter, let's change the treatment of the problem. First, we must have an expression for the $\textit{density matrix}$ of the system; but, we cannot build it before change the basis to a new one in order to not assume the separability of the wave function. 

In the Eq. (1) we have an expression of an arbitrary state, let's get the density matrix for an arbitrary state:

\begin{equation}
        \begin{split}
                \rho & = \ket{\psi}\bra{\psi}
        \end{split}
\end{equation}

Now, we will use the vector representation for the spin-1/2 particles: $\ket{\uparrow} \equiv \left( \begin{matrix} 1 \\ 0 \end{matrix} \right) $ and $\ket{\downarrow} \equiv \left( \begin{matrix} 0 \\ 1 \end{matrix} \right) $


\begin{equation}
        \begin{split}
                \rho & = \left[a_1 \left( \begin{matrix} 0 \\ 1 \end{matrix} \right) + a_2 \left( \begin{matrix} 1 \\ 0 \end{matrix} \right) \right] \left[ a_1^* \left( \begin{matrix} 0 & 1 \end{matrix} \right) + a_2^* \left( \begin{matrix} 1 & 0 \end{matrix} \right) \right] \\
                & = \left( \begin{matrix} |a_1|^2 & a_1a_2^* \\ a_2a_1^* & |a_2|^2 \end{matrix} \right)
        \end{split}
\end{equation}

Turns out that, because we have a 2x2 matrix, and should be hermitian ($\rho = \rho^{\dagger}$), then we can decompose this matrix in the Pauli and Identity matrices basis in the form:

\begin{equation}
        \begin{split}
                \rho & = \sum_{i=1}^{3} c_i \sigma_i \\
                & = c_0 \mathcal{I} + c_x \sigma_x + c_y \sigma_y + c_z \sigma_z \\
                & = \left( \begin{matrix} c_0 + c_z & c_x -ic_y \\ c_x +ic_y & c_0 -c_z \end{matrix} \right) 
        \end{split}
\end{equation}

With this in mind, note $\rho = \rho^{\dagger}$ and also remember that Tr($\rho$) = 1, so:
\begin{equation}
        2c_0 = 1 \longrightarrow c_0 = \frac{1}{2}
\end{equation}

So, now we have an expression for the density matrix:

\begin{equation}
        \rho = \frac{1}{2}\left[ I + \vec{P}\cdot \vec{\sigma_k} \right]
\end{equation}

where we say that $\vec{P}$ is the $\textbf{Polarization vector}$ and $\sigma_k = (\sigma_x, \sigma_y, \sigma_z)$. 

\begin{equation}
        \begin{split}
                \sigma_x & = \left( \begin{matrix} 0 & 1 \\ 1 & 0 \end{matrix} \right) \\
                \sigma_y & = \left( \begin{matrix} 0 & -i \\ i & 0 \end{matrix} \right) \\
                \sigma_z & = \left( \begin{matrix} 1 & 0 \\ 0 & -1 \end{matrix} \right)
        \end{split}
\end{equation}

Now, we are in position to express the $\textbf{\textit{Polarization density matrix}}$ in the Pauli matrix basis as:

\begin{equation}
        \begin{split}
                \rho & = \left( \ket{\psi_1} \otimes \ket{\psi_2} \right) \left(\bra{\psi_1} \otimes \bra{\psi_2} \right) \\
                & = \ket{\psi_1} \bra{\psi_1} \otimes \ket{\psi_2}\bra{\psi_2} \\
                & = \frac{1}{2} \left[ \mathcal{I} + \vec{P} \cdot \vec{\sigma} \right] \otimes \frac{1}{2} \left[ \mathcal{I} + \vec{Q}\cdot \vec{\sigma} \right] \\
                & = \frac{1}{4} \left[ \mathcal{I} \otimes \mathcal{I} + \vec{P} \cdot \vec{\sigma} \otimes \mathcal{I} + \mathcal{I} \otimes \vec{Q} \cdot \vec{\sigma} + \vec{P} \cdot \vec{\sigma} \otimes \vec{Q} \cdot \vec{\sigma} \right] \\
                & = \frac{1}{4} \left[ \mathcal{I_2} \otimes \mathcal{I_2} + \sum_{i=1}^{3} B_i^+ (\sigma_i \otimes \mathcal{I_2}) + \sum_{j=1}^3 B_j^- (\mathcal{I_2} \otimes \sigma_j ) + \sum_{i, j = 1}^3 C_{ij} (\sigma_i \otimes \sigma_j ) \right]
        \end{split}
\end{equation}

In this last expression, we have defined: 

\begin{equation*}
        \begin{split}
                B_i^+ & : \text{Fermion 1 polarization   } \vec{P} \\
                B_j^- & : \text{Fermion 2 polarization   } \vec{Q} \\
                C_{ij} & : \text{Polarizations product   } \vec{P} \cdot \vec{Q}
        \end{split}
\end{equation*}

In the following, we'll be using $\textbf{Sympy}$ and $\textbf{Numpy}$ libraries for developing a calculation tool for the expression in Eq. (10). 

First, we define the variables that are present in the PDM expression: 

```python
import sympy as sy
import numpy as np

#Definitions for sigma matrix and identity matrix
id = np.array([[1, 0], [0, 1]])
sigma_1 = np.array([[0, 1], [1, 0]])
sigma_2 = np.array([[0, -sy.I], [sy.I, 0]]) #using the sympy notation for i (cleaner)
sigma_3 = np.array([[1, 0], [0, -1]])
#All sigma in one array
sigmas = [sigma_1, sigma_2, sigma_3]

#variables B_i +- y C_ij
Bp = np.array(sy.symbols('Bp1:4')) 
Bm = np.array(sy.symbols('Bm1:4'))
Cij = np.array([sy.symbols('C_1:4_1:4')]).reshape(3,3) #Matrix 3x3
```

Now, we start to realize the computation for the expressions. The total PDM will be take in four parts in order to get a better (and cleaner) view in the calculation. 

```python
#Kronecker product for identities (part 1)
part1 = np.kron(id, id)

#First sum on density matrix (parte 2)
part2_aux = np.array([Bp[i] * np.kron(sigmas[i], id) for i in range(3)])
part2 = sum(part2_aux[i] for i in range(3))

#Second sum on density matrix (parte 3)
part3_aux = np.array([Bm[j] * np.kron(id, sigmas[j]) for j in range(3)])
part3 = sum(part3_aux[j] for j in range(3))

#Third sum on density matrix (parte 4)
part4_aux = np.array([Cij[k][l] * np.kron(sigmas[k], sigmas[l]) for k in range(3) for l in range(3)])
part4 = sum(part4_aux[i] for i in range(9))

#Sum of all contributions (all parts)
rho = (part1 + part2 + part3 + part4 ) #Here we need to implement the 1/4 factor, but for a cleaner output we ignore them until we need it to perform correct calculations
```

Once we have all the expression for the PDM in the $\texttt{rho}$ variable, the output will give us all the necessary details about the density matrix:

```python
print("Full Density matrix (there is missing the 1/4 for cleaner output):")
print(rho)

#Trace of density matrix:
trace = sum(rho[i][i] for i in range(4))
print("Trace rho:", trace / 4)

#Transpose density matrix:
print("Traspose density matrix:")
print(rho.T)

#Verificando unitariedad:
if rho.all() == rho.T.all():
    print("Density matrix is the same as the transpose")
```

# Entanglement and Bell inequality violation

Given a two-qubit, 4x4 density matrix $\rho$ as in the upper step, its concurrence can be explicitly constructed by using the auxiliary matrix:

\begin{equation}
    R = \rho (\sigma_y \otimes \sigma_y) \rho^* (\sigma_y \otimes \sigma_y) 
\end{equation}

where $\rho^*$ denotes the matrix with complex conjugated entries. In the following, we intend to calculate this $R$ matrix and see the eigenvalues. So, we define the $\rho^*$ matrix and $R$:

```python
rho = rho / 4 #Now, because we need to get a correct eigenvalue calculation, we now add the 1/4 factor
rho_conj = np.conj(rho)
aux_1 = np.kron(sigmas[1], sigmas[1])
R = rho * aux_1 * rho_conj * aux_1 
R_matrix = sy.Matrix(R)
print("R=", sy.latex(R_matrix))
```

$ Z_{1} = \left( \frac{C_{1 1}}{4} - \frac{i C_{1 2}}{4} - \frac{i C_{2 1}}{4} - \frac{C_{2 2}}{4}\right) $
